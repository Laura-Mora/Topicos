```python
def gender_features(word): 
    return {'last_letter': word[-1:]} 

print("Test gender_features: Shrek - k")
gender_features('lorena') 
```

```python
print(len(names.words('male.txt')),names.words('male.txt')[:20])
print(len(names.words('female.txt')),names.words('female.txt')[:20])
```

```python
names = ([(name, 'hombre') for name in names.words('male.txt')] + [(name, 'mujer') for name in names.words('female.txt')]) 
```

```python
print(len(names), names)
```

```python
random.shuffle(names) 
```

```python
print(len(names), names)
```

```python
featuresets = [(gender_features(n), g) for (n,g) in names] 
print(featuresets)
train_set, test_set = featuresets[1000:], featuresets[:1000]
#print("train_set :",[train_set[0],train_set[1],train_set[3]])
print("test_set :", test_set)
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set) 
print(classifier)
```

```python
print (nltk.classify.accuracy(classifier, test_set) )

print(classifier.classify(gender_features('carolina')) )
print(classifier.classify(gender_features('gabriel')) )
print(classifier.classify(gender_features('andres')))
print(classifier.classify(gender_features('andrea')) )
print(classifier.classify(gender_features('fabian')) )
```
