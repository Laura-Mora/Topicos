```
import twitter
import networkx as nx
import os
from twitter import Api
from datetime import datetime
```

```
graph_followers =nx.DiGraph()
graph_friends =nx.DiGraph()
graph_all = nx.DiGraph()
```

```
date_file_name = datetime.now().strftime('%Y-%m-%d_%H-%M')
```

```
username = "gabrielmoreno10"

conf = {
 'CONSUMER_KEY': '',
 'CONSUMER_SECRET': '',
 'ACCESS_TOKEN': '',
 'ACCESS_TOKEN_SECRET': ''}

api = Api(conf['CONSUMER_KEY'], conf['CONSUMER_SECRET'], conf['ACCESS_TOKEN'], conf['ACCESS_TOKEN_SECRET'])
```

```
# Load data
username = "gabrielmoreno10"
print ("Loading users who follow", username, "...")
followers = []
followers.extend(api.GetFollowers(screen_name=username))
print(followers)
print ("Loading users that", username,"follows...")
friends = []
friends.extend(api.GetFriends(screen_name=username))
print(friends)
```

```
print(followers)
print(friends)
```

```
print('followers: ', followers[:5])
print('friends: ', friends[:5])
```

```
print(followers[0])
```

```
# Create graph
print ("Adding followers relationships...")
for user in followers:
	graph_followers.add_edge(user.name, username)
	
print ("Adding following relationships...")
for user in friends:
	graph_friends.add_edge(username, user.name)
 
# Create graph
print ("Adding followers relationships...")
for user in followers:
	graph_all.add_edge(user.name, username)
	
print ("Adding following relationships...")
for user in friends:
	graph_all.add_edge(username, user.name)
```

```
# Save graph
print ("The personal profile was analyzed succesfully.")
nx.draw(graph_followers.to_directed())
```

```
# Save graph
print ("The personal profile was analyzed succesfully.")
nx.draw(graph_friends.to_directed())
```

## Section 2:
We are going to connect to twitter, and we will get a search from tokens.

```
def extract_hash_tags(s):
  return set(part[1:] for part in s.split() if part.startswith('#'))
def extract_mention(s):
  return set(part[1:] for part in s.split() if part.startswith('@'))
```

```
hashtag = 'reforma tributaria'
query = api.GetSearch(term=hashtag, count=100)
graph=nx.DiGraph()

print('Result Query: ', query)
print ("Found: ",len(query)," results.")
```

```
print(query[90])
```

```
# Get results and find retweets and mentions
count = 0
for result_raw in query:  
  result = result_raw.AsDict()
  print ("Tweet:",result)
  print ("By user:",result["user"]["name"])
  if 'user_mentions' in result:
    print ("Tweet Number :", count)
    count += 1
    if len(result["user_mentions"]) > 0:
      print ("Mentions:")
      for i in result["user_mentions"]:
        print(i)
        print (" - by", i["screen_name"])
        graph.add_edge(i["screen_name"], result["user"]["name"])
    if "retweeted_status" in result:
      if len(result["retweeted_status"]["user_mentions"]) > 0:
        print ("Retweets:")
        for i in result["retweeted_status"]["user_mentions"]:
          print (" - by",i["screen_name"])
          graph.add_edge(i["screen_name"],result["user"]["name"])
    else:
      pass
```

```
# Save graph
print("The network of RT of the hashtag was analyzed succesfully.")
print("Saving the file as " + hashtag)
nx.draw(graph, with_labels=True)
```

```
print('Nodes :', len(graph.nodes))
print('Edges :', len(graph.edges))
```

```
print('all_degree', graph.degree())
print('in_degree', graph.in_degree())
print('out_degree', graph.out_degree())

list_data = list(graph.degree())
print(list_data)
print(list_data.sort(key=lambda tup: tup[1]))
#sorted(graph.degree())
```

```
# print('all_degree', graph.degree())
# print('in_degree', graph.in_degree())
# print('out_degree', graph.out_degree()
from operator import itemgetter 
popular_people = [{'person':person, 'value':graph.degree(person)} for person in graph.nodes]
print(popular_people)
list_central = sorted(popular_people, key=itemgetter('value'), reverse=True)
print(sorted(popular_people, key=itemgetter('value'), reverse=True))
```
