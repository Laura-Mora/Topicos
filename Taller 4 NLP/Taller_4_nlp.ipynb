```python
stopwrds = list(set(stopwords.words('spanish') ))
```

```python
def tag_words(words):
    return dict([(word, True) for word in words.split() if word not in stopwrds])
```

## Clasificar topic

Explorando los datos de COlombia, encontramos que en parte de los registros, el texto contiene el nombre del tópico. Lo cual, en una primera prueba del modelo, generó un overfitting bastante notable.

Por esto mismo, procedimos a crear una función que elimine las palabras de los tópicos y se aplique a todos los textos del dataframe.

```python
#!pip install unidecode
```

```python
import unidecode
```

```python
list(df_col.topic.unique())
```

```python
def drop_topic_name(text):
  text = unidecode.unidecode(text)
  topics = list(df_col.topic.unique())
  u_topics = [x.upper() for x in topics]
  topics = topics + u_topics
  for word in topics:
    if word in text:
      text = text.replace(word, "")
  return text
```

```python
df_col["text_e"] = df_col.text.apply(lambda x: drop_topic_name(x))
```

```python
df_col["text_e"][0]
```

```python
df_col["topic"].value_counts()
```

```python
df_col.head()
```

```python
df_col_to_train = [(tag_words(df_col.at[i, "text_e"]), df_col.at[i, "topic"]) for i in range(len(df_col))]
```

```python
train_set, test_set = df_col_to_train[int(len(df_col_to_train)*0.7):], df_col_to_train[:int(len(df_col_to_train)*0.7)]
```

```python
classifier_topics = nltk.NaiveBayesClassifier.train(train_set) 
print(classifier_topics)
```

```python
print(nltk.classify.accuracy(classifier_topics, test_set))
```

```python
print(nltk.classify.accuracy(classifier_topics, train_set))
```

```python
data_filename_topics = f"file_classifier_topics_{datetime.now().strftime('%Y-%m-%d')}.vec"
dump(classifier_topics, data_filename_topics)
```
